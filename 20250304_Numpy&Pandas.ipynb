```python
points_mean = winemag.points.mean()
```

```python
def remean_point(point):
  point = point - points_mean
  return point
```

```python
winemag.points.map(remean_point)
```

```python
winemag.points.map(lambda p: p - points_mean)
```

```python
def remean_point(row):
  row.points = row.points = points_mean
  return row

winemag.apply(remean_point, axis='columns')
```

```python
def sum_column(points):
  return sum(points)

winemag.groupby(['country','province']).points.agg([len, min, max, sum_column])
```

```python
newdata = winemag.groupby(['country','province']).points.agg([len, min, max, sum_column])
```
